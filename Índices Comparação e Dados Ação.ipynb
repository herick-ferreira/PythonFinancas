```
ibov = inv.get_index_historical_data('Bovespa', country='brazil', from_date='01/01/1900', to_date='19/02/2021')
ibov
```

Índice IFIX

```
ifix = inv.get_index_historical_data('BM&FBOVESPA Real Estate IFIX', country='brazil', from_date='01/01/1900', to_date='19/02/2021')
ifix
```

Comparando os Índices

```
indices = pd.DataFrame()
data_inicio = ifix.index[0]
 
indices['ifix'] = ifix['Close'] / ifix['Close'].loc[data_inicio]
indices['ibov'] = ibov['Close'] / ibov['Close'].loc[data_inicio]
```

```
indices.plot()
```
