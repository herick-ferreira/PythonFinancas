 Buscando dados no mercado

```
ibov = web.get_data_yahoo('^BVSP')
```

```
ibov.head()
```

```
ibov.tail()
```

```
ibov["Close"].plot(figsize=(22,8), label="IBOV")
ibov["Close"].rolling(21).mean().plot(label="MM21")
ibov["Close"].rolling(200).mean().plot(label="MM200")
plt.legend()
```

```
 
ibov_periodo = ibov[ibov.index.year == 2008]
 
ibov_periodo["Close"].plot(figsize=(22,8), label="IBOV")
ibov_periodo["Close"].rolling(21).mean().plot(label="MM21")
ibov_periodo["Close"].rolling(200).mean().plot(label="MM200")
plt.legend()
```

```
ibov_periodo = ibov[(ibov.index.year >= 2008) & (ibov.index.year <= 2014)]
 
ibov_periodo["Close"].plot(figsize=(22,8), label="IBOV")
ibov_periodo["Close"].rolling(21).mean().plot(label="MM21")
ibov_periodo["Close"].rolling(200).mean().plot(label="MM200")
plt.legend()
```
