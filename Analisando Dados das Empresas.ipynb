# **Importando Bibliotecas**

```
import numpy as np
import pandas as pd
import string
import warnings
warnings.filterwarnings('ignore')
 
import requests
```

# **Obtendo Dados das Empresas**

```
url = 'https://www.fundamentus.com.br/resultado.php'
```

```
header = {
  "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36"  
 }
 
r = requests.get(url, headers=header)
```

```
df = pd.read_html(r.text,  decimal=',', thousands='.')[0]
```

```
df
```

# **Analisando os dados**

```
##Empresas com Liquidez acima de R$1.000.000

df = df[df['Liq.2meses'] > 1000000]
```

###Melhores Empresas Segundo a Magic Fómula Joel Greenblatt

EV/EBIT acima de 0 ordenados pelos maiores EV/EBIT e ROIC

```
ranking = pd.DataFrame()
ranking['pos'] = range(1,151)
ranking['EV/EBIT'] = df[ df['EV/EBIT'] > 0 ].sort_values(by=['EV/EBIT'])['Papel'][:150].values
ranking['ROIC'] = df.sort_values(by=['ROIC'], ascending=False)['Papel'][:150].values
```

```
ranking
```

Ranking das Empresas  com os Melhores Dados

```
a = ranking.pivot_table(columns='EV/EBIT', values='pos')
```

```
b = ranking.pivot_table(columns='ROIC', values='pos')
```

```
t = pd.concat([a,b])
t
```

A Soma dos Dois Indicadores

```
rank = t.dropna(axis=1).sum()
rank
```
